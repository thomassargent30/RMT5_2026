```python
import numpy as np
import numpy.linalg as LA
import matplotlib.pyplot as plt
from quantecon import LQ
```

```python
a0 = 10.

a1 = 1.

β = 0.99

ρ0 = 0. 

ρ1 = 1.2

ρ2 = -0.4

d = 0.1

σ = 0.1
```

```python
R = np.matrix(
    [[-a1, a0/2, -1/2, 0.],
     [a0/2, 0., 0., 0.],
     [-1/2, 0., 0., 0.],
     [0., 0., 0., 0.]])

A = np.matrix(
    [[1., 0., 0., 0.],
     [0., 1., 0., 0.],
     [0., ρ0, ρ1, ρ2],
     [0., 0., 1., 0.]])

B = np.matrix([[1.], [0.], [0.], [0.]])

C = np.matrix([[0.], [0.], [σ], [0.]])

Q = np.matrix(d)
```

```python
lq = LQ(Q, R, A, B, C, beta = β, T = None, Rf = None)
```

```python
P, F, h = lq.stationary_values()
```

```python
def verify1(A, B, Q, P, F):
    
    n, n = np.shape(P)
    I = np.identity(n)
    lhs = LA.inv(I + B @ LA.inv(Q) @ B.T @ P) @ A
    rhs = A - B @ F
    #print(lhs, rhs)
    
    if (lhs == rhs).all:
        print('True')
    else:
        print ('False')
```

```python
verify1(A, B, Q, P, F)
```

```python
def verify2(A, B, Q, P, F):
    
   
    rhs = (R + F.T @ Q @ F) + (A - B @ F).T @ P @ (A - B @ F)
    lhs = P
    
    #print(lhs, rhs)
    if (lhs == rhs).all:
        print('True')
    else:
        print ('False')
```

```python
verify2(A, B, Q, P, F)
```

```python
def verify3(A, B, Q, P, F):
    
   
    rhs = R + A.T @ P @ A
    lhs = P
    
    #print(lhs, rhs)
    if (lhs == rhs).all:
        print('True')
    else:
        print ('False')
```

```python
verify3(A, B, Q, P, F)
```
