```python
import re
import os
import pandas as pd
```

```python
def from_color_to_gray(file_name,input_file_loc="",output_file_loc=""):
    """
    function:
    This function will change the colorful picture file_name.eps which located 
    in input_file_loc into gray picture file_name_gray.eps which located 
    output_file_loc.
    The default setting will just output generated file from/into current folder
    
    parameter:
    input_file_loc: the file location of colorful eps which we plan to change. 
    If this file does not exist before, we will create this file.
    The default setting will just import colorful picture from current folder

    output_file_loc: the file location of gray eps which we plan to generate
    The default setting will just output generated file into current folder

    file_name: the name of file which we want to change, there is no need to add .eps
    
    """
    if len(output_file_loc)>0:
        import os
        try:
            os.mkdir(output_file_loc)
        except:
            pass
      
    rgb_hunter = re.compile("([0-9]+\.*[0-9]*)\s+([0-9]+.*[0-9]*)\s+([0-9]+.*[0-9]*)\s+sr")

    replace_str = "{gray} sg"

    fname = file_name+".eps"
    outfname = file_name+"_gray.eps"

    with open(input_file_loc+fname, "r") as old_f:
        with open(output_file_loc+outfname, "w") as new_f:
            for line in old_f.readlines():
                hunted = rgb_hunter.search(line)
                if hunted is None:
                    new_f.write(line)
                else:
                    s, e = hunted.span()
                    r, g, b = hunted.groups()
                    r = float(r)
                    g = float(g)
                    b = float(b)
                    gray = 0.299*r + 0.587*g + 0.114*b
                    newline = line.replace(line[s:e], replace_str.format(gray=gray))
                    new_f.write(newline)
```
