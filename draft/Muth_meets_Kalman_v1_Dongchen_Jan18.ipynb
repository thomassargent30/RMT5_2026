# Muth and Kalman

## 1. Innovations representation

 Consider the linear state space system
 
$$\eqalign{ x_{t+1} & = \rho x_t + c w_{t+1} \cr
            y_t & = x_t + v_t , } \  \ (1) $$
            
where $ \rho \in [0, 1]$ and $c$ are scalars, $\{w_t\}_{t=0}^\infty$ is an i.i.d. scalar process with $w_t \sim N(0,1)$,
 $\{v_t\}_{t=0}^\infty $ is an i.i.d. process of measurement errors with $E v_t w_{s+1} = 0$ for all $t\geq 0 $ and $s\geq 0 $ with
 $v_t \sim N(0,R)$, and $x_0$ is a random initial value distributed as $x_0 \sim N(0, \frac{c^2}{1-\rho^2})$. ( Note: Muth (1960)
 considered the limiting case in which $\rho \uparrow  1$.)



* The univariate process  $\{x_t\}_{t=0}^\infty$ is  Markov.


*  The univariate process $\{y_t\}_{t=0}^\infty$ is not Markov?

* We construct an **innovations representation** for $\{y_t\}_{t=-\infty}^\infty$ of the time-invariant form

$$\eqalign{ \hat x_{t+1} & = \rho \hat x_t + k a_t \cr
            y_t & = \hat x_t + a_t,  } \  \ (2) $$
            
where $k$ is a regression coefficient,  $\hat x_t = E x_t | y^{t-1}$, $a_t = y_t - E [y_t | y^{t-1}] $,  $a_t$ is by construction serially uncorrelated and homoskedastic
with $E [a_t| y^{t-1}] = 0$ and $E a_t^2 = \omega$, and $y^{t}$ is the semi-infinite history $y_t, y_{t-1}, \ldots$.
$k$ is the regression coefficient of $x_t$ or $x_t \hat x_t$  on $a_t$. 


* The Kalman filter supplies the following formulas for $k$ and $\omega$: 

    * Let $Q = c^2$ and let $\Sigma$ be the positive fixed point of the algebraic Ricatti equation, namely,
    $$ \Sigma = \frac{ [ Q + (\rho^2 -1) R] + \sqrt{[ Q + (\rho^2 -1) R]^2 + 4 Q R}}{2} $$ 
    Then $$\omega = \Sigma + R $$ and 
    $$ k = \frac{\rho \Sigma}{\Sigma + R} $$
    
    
Please make the default values be $c=1, R =1 , \rho=1$.   Remember that $Q = c^2$.

(Don't worry about this making the initial distribution for $x_0$ have infinite  variance. I'll explain.)

### Compute $\Sigma, \omega$ and $k$ with Python

We use Python to compute the fixed point $\Sigma$ as well as 
the one-step ahead prediction error variance $\omega$ and
the regression coefficient $k$:

```python
import numpy as np


# Set parameters
c = 1
R = 1
ρ = 1
Q = c**2


# Define functions
sqrt = lambda x: np.sqrt(x)


# Compute Σ
numerator = (Q + (ρ**2 - 1) * R) + sqrt((Q + (ρ**2 - 1) * R)**2 + 4 * Q * R)
Σ = numerator / 2


# Compute ω
ω = Σ + R


# Compute k
k = ρ * Σ / (Σ + R)


# Print results
print("Σ: {:6.5f}".format(Σ))
print("ω: {:6.5f}".format(ω))
print("k: {:6.5f}".format(k))
```

We also want to check that the above fixed point for $\Sigma$ is the limit of iterations on
    
$$ \Sigma_{t+1} = \rho^2 \Sigma_t + Q - \frac{\rho^2 \Sigma_t^2}{\Sigma_t + R} $$
    
starting from any $\Sigma_0 > 0$.

```python
# Set initial value
Σ_0 = 1

# Set tolerance and maximum iterations
tol = 1e-16
max_iter = 10000


# Iterate
i = 0

while i < max_iter:
    i += 1
    
    # Compute Σ_{t+1}
    Σ_1 = ρ**2 * Σ_0 + Q - (ρ * Σ_0)**2 / (Σ_0 + R)
    if abs(Σ_1 - Σ_0)>tol:
        Σ_0 = Σ_1
        if i==max_iter:
            print("Σ is not converged in the maximum iteration of {}".format(max_iter))
    else:
        break
        
print("Is Σ_t equal to Σ?")
print(Σ_0 == Σ)
    
```

## 2. Autoregressive representation

We have the following formulas:

$$ \eqalign{ \hat x_{t+1} & = k \sum_{j=0}^\infty (\rho - k)^j y_{t-j}  \cr
             y_t & = k \sum_{j=0}^\infty (\rho - k)^j y_{t-j-1} + a_t , }$$
             
where $a_t $ is orthogonal to $y_{t-j-1}$ for $j \geq 1$ so that the second equation is an **autoregressive representation** or **autoregression**
for $y_t$.  It can be  verified that $E[ y_{t+1} | y^t] = \hat x_{t+1}$.


When $v^t$ denotes a semi-infinite history of a scalar process $v_t$, we have the following formulas:

$$ \eqalign{ x_t & = c \sum_{j=0}^\infty \rho^j w_{t-j} \cr
             y_t & =   c \sum_{j=0}^\infty \rho^j w_{t-j} + v_t \cr
             E[y_t | w^{t-1}, v^{t-1}] & = c \sum_{j=1}^\infty \rho^j w_{t-j} \cr
             E ( y_t - [y_t | w^{t-1}, v^{t-1}])^2 & = c^2 + R.}$$
             
We can verify that the one-step ahead prediction error variance    $E ( y_t - [y_t | w^{t-1}, v^{t-1}])^2$ is less than
the one-step ahead prediction error variance $E a_t^2$.  To do this simply
verify that

$$ Q + R < \Sigma+ R $$


```python
print((Q+R) < (Σ+R))
```

When $L$ is the lag operator  we have  the following formula:

$$ a_t = {\frac{1}{1- (\rho-k)L}} \left[ cw_t + (1-\rho L) v_t \right] $$

or 

$$ a_t = c \sum_{j=0}^\infty (\rho - k)^j w_{t-j} + v_t - k \sum_{j=1}^\infty (\rho-k)^j v_{t-j} $$

or 

$$ a_t = c \sum_{j=0}^\infty (\rho - k)^j w_{t-j} + \sum_{j=0} h_j v_{t-j} $$

where 

$$ h_j = \cases{ 1 & if $j = 0$ \cr
                 -k (\rho -k )^{j-1} & if $j \geq 1 $ } $$

This formula  shows how the contemporaneous scalar shock $a_t$ reflects *old news* in the form of past
$w_t$s and past $v_t$s.  

Note how the discrepancy in variances of one-step ahead prediction errors that you computed above makes sense.



### Compute and plot the cutocovariance functions with Python

We use Python to accomplish the following tasks:
* to compute and plot the autocovariance function of $c \sum_{j=0}^\infty (\rho - k)^j w_{t-j}$ which
equals
 

$$\gamma(j) = c^2 (\rho-k)^j, \quad  j \geq 0$$

  
* to compute and plot the autocovariance function of $v_t - k \sum_{j=1}^\infty (\rho-k)^j v_{t-j}$ which
equals

  $$ \tilde \gamma(j) = \cases{ R & if $j =0$ \cr
                     - R k (\rho -k )^{j-1} & if $j \geq 1$ }$$
                     
                     
* to compute and plot the sum of the autocovariance functions $\gamma(j) $ and $\tilde \gamma(j)$
as a function of $j = 0, 1 \ldots$.


```python
import matplotlib.pyplot as plt


# Setup
γ = []
γ̃ = []
sum_auto = []
max_j = 10
j_space = np.arange(0,max_j+1)


# Compute γ, γ̃ and the sum
for j in j_space:
    
    # Compute γ
    gamma = c**2 * (ρ - k)**j
    γ.append(gamma)
       
    # Compute γ̃
    if j == 0:
        gamma_tilde = R
    else:
        gamma_tilde = -R * k * (ρ - k)**(j - 1)
    γ̃.append(gamma_tilde)

    # Compute sum
    sum_auto.append(gamma + gamma_tilde)


# Plot
fig, ax = plt.subplots(3, 1, figsize=(12,12))

# γ 
ax[0].plot(j_space, γ, color="k", linestyle="-")
ax[0].set_title(r"$\gamma$")
ax[0].set_xlim([0, max_j])
ax[0].set_ylim([-0.7, 2.0])
ax[0].grid()

# γ̃ 
ax[1].plot(j_space, γ̃, color="k", linestyle=":")
ax[1].set_title(r"$\tilde{\gamma}$")
ax[1].set_xlim([0, max_j])
ax[1].set_ylim([-0.7, 2.0])
ax[1].grid()

# the sum
ax[2].plot(j_space, sum_auto, color="k", linestyle="-.")
ax[2].set_title(r"autocovariance of $a_t$")
ax[2].set_xlim([0, max_j])
ax[2].set_ylim([-0.7, 2.0])
ax[2].set_xlabel(r"$j$")
ax[2].grid()


plt.show()

```
