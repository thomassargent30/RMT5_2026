```python
import igraph
from network2tikz import plot
```

# 0 Overview

All tikz graphs are composed by two components: node(s) and edge(s).

Here we are going to draw three simple tikz graphs:
- the benchmark is a graph of two nodes and one edge, see section ``1 two nodes and one edge``,
- in section ``2 two nodes and one edge with an edge label``, we will add a label to the edge in the benchmark graph from section ``1 two nodes and one edge``,
- in section ``3 two nodes and one edge with two node labels``, we will add two different labels to the nodes in the benchmark graph from section ``1 two nodes and one edge``.

Let's get started.

# 1 two nodes and one edge

Here we are going to draw two nodes, ``node 0`` and ``node 1``, and a directed edge from ``node 0`` to ``node 1``.

(**Step 1**) First we define the edge and its direction:
- a pair ``(i, j)`` defines the edge from node ``i`` to node ``j``, where $i, j \in \{0, 1, 2, \cdots N\}, N \in \mathbb N$, and
- let ``directed=True`` if you want an arrow at the end of the edge; otherwise, let ``directed=False``.

```python
net = igraph.Graph([(0,1)],directed=True)
```

(**Step 2**) Then we define the number of nodes the their locations.

(Defining the location of nodes is a bit tricky and you need to try and see the difference.)

```python
layout = {0: (5, -4), 1: (7, -4)}
```

(**Step 3**) Next we can define the styles or properties we want for the nodes and edges. 

(Here we only define the location of nodes, but we can do a lot more than that. For more, please see [here](https://github.com/hackl/tikz-network/blob/master/manual.pdf).)

```python
visual_style1 = {}
visual_style1['layout'] = layout
```

(**Step 4**) Finally we can use the method ``plot`` from ``network2tikz`` to plot the tikz graph we defined above.

(We can generate tikz graphs in many formats by simply replacing ``xxx.pdf`` with ``xxx.csv``, etc.)

```python
plot(net, 'simple_tikz_graph1.pdf', **visual_style1)
```

# 2 two nodes and one edge with an edge label

Now we want to add a label to the edge of above tikz graph generated in section ``# 1 two nodes and one edge``.

Ceteris Paribus, we only need to 
1. add defining the label of the edge in **Step 2** and

(Note that we use ``net.es[]`` here, if we want to define property for nodes, we use ``net.vs[]``.)

```python
net.es['label'] = ['100'] # here the edge label is '100'
```

2. specify the ``edge_label`` in **Step 3**

```python
visual_style2 = {}
visual_style2['layout'] = layout
visual_style2['edge_label'] = net.es['label']
```

Finally we plot it.

```python
plot(net, 'simple_tikz_graph2.pdf', **visual_style2)
```

# 3 two nodes and one edge with two node labels

Lastly we want to add two different labels to the nodes of above tikz graph generated in section ``# 1 two nodes and one edge``.

Ceteris Paribus, we only need to 
1. add defining the label of the nodes in **Step 2** and

```python
net.vs['label'] = ['0', '1'] # here the node labels are '0' and '1'
```

2. specify the ``vertex_label`` in **Step 3**

```python
visual_style3 = {}
visual_style3['layout'] = layout
visual_style3['vertex_label'] = net.vs['label']
```

Finally we plot it.

```python
plot(net, 'simple_tikz_graph3.pdf', **visual_style3)
```
